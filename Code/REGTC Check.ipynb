```python
import pandas as pd
```

```python
DF=pd.read_csv(r'cthrr460_14449656.csv', skiprows=10)
```

```python
DF.head()
```

```python
def clean_data(DF):
    # Filter rows based on column: 'Employee Name'
    DF = DF[~DF['Employee Name'].str.contains("------", regex=False, na=False, case=False)]
    # Replace gaps forward from the previous valid value in: 'Employee Name', 'EmplID', 'Empl'
    DF = DF.fillna({'Employee Name': DF['Employee Name'].ffill(), 'EmplID': DF['EmplID'].ffill(), 'Empl': DF['Empl'].ffill()})
    return DF

DF_clean = clean_data(DF.copy())
DF_clean.head()
```
